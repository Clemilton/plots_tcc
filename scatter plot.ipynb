```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['savefig.dpi'] = 120
plt.rcParams['figure.dpi'] = 120
```

```python

x=np.linspace(0,1,40)
y=x
df = pd.read_csv('mlp_fcn.csv')
name = df['Name'].values
mlp = df['MLP'].values
fcn = df['FCN'].values
mlp=mlp/100
fcn=fcn/100

fig,ax = plt.subplots()

ax.fill_between(x, y, 1,facecolor='orange', alpha=0.3)
plt.plot(x,y,'-',linewidth=1,color='orange')
plt.scatter(mlp, fcn, s=35, alpha=0.5,color='blue',edgecolors='black')
for i,txt in enumerate(name):
    ax.annotate(txt,(mlp[i], fcn[i]))
plt.xlabel('MLP')
plt.ylabel('FCN')
plt.axis([0,1,0,1])
plt.show()
```
